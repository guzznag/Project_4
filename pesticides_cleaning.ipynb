```python
import pandas as pd
from pathlib import Path
```

```python
pesticides_df = pd.read_csv(Path("2019_pesticide_by_county.csv"))
pesticides_df.head()
```

```python
pesticides_df = pesticides_df.fillna(0)
pesticides_df.head()
```

```python
pesticides_df.info
```

```python
pesticides_df.columns
```

```python
pesticides_df = pesticides_df.sort_values("Grand Total", ascending = False)
pesticides_df = pesticides_df.drop(columns='Grand Total')
pesticides_df.head()
```

```python
types_df = pd.read_csv("pesticide_types.csv")
types_df.head()
```

```python
pesticides_df = pd.concat([pesticides_df, types_df], ignore_index=False)
pesticides_df = pesticides_df.drop(columns="Name")
pesticides_df.tail()
```
