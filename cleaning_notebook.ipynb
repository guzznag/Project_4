```python
import pandas as pd
from pathlib import Path
```

```python
#calling in the 2019 pesticides table
pesticides_df = pd.read_csv(Path("2019_pesticide_by_county.csv"))
pesticides_df.head()
```

```python
#filling in the null values with a 0
pesticides_df = pesticides_df.fillna(0)
pesticides_df.head()
```

```python
#looking at each of the different column names/dataframe info
pesticides_df.info
```

```python
#looking at each column name
pesticides_df.columns
```

```python
#dropping the grand total column since we don't really need it
pesticides_df = pesticides_df.drop(columns = "Grand Total")
pesticides_df
```

```python
#creatng variables to hold specific columns that fall within that category 
insecticides = pesticides_df[["ACEPHATE", "ACETAMIPRID", "CARBARYL","CHLORANTRANILIPROLE",
                              "CHLORPYRIFOS","CLOTHIANIDIN","CYANTRANILIPROLE",
                              "DIAZINON","DICROTOPHOS","DIFLUBENZURON","DIMETHOATE",
                              "DINOTEFURAN","ETHOPROPHOS","ETOXAZOLE","FIPRONIL",
                              "IMIDACLOPRID","MALATHION","METHOMYL","METHOXYFENOZIDE",
                              "PERMETHRIN","SULFOXAFLOR","TEBUPIRIMPHOS","TERBUFOS",
                              "THIAMETHOXAM"]]
herbicides = pesticides_df[["2,4-D","ACETOCHLOR","ATRAZINE","BENTAZONE",
                            "BROMACIL","BROMOXYNIL","CHLORIMURON","DICAMBA",
                            "DIMETHENAMID","DIMETHENAMID & DIMETHENAMID-P",
                            "DIMETHENAMID-P","DIURON","FLUOMETURON","GLYPHOSATE",
                           "HALOSULFURON","HEXAZINONE","IMAZETHAPYR","LINURON","METOLACHLOR",
                           "METOLACHLOR & METOLACHLOR-S","METOLACHLOR-S","METRIBUZIN",
                           "ORYZALIN","PROPAZINE","SIMAZINE","SULFENTRAZONE","TEBUTHIURON",
                           "THIOBENCARB","TRICLOPYR"]]
fungicides = pesticides_df[["AZOXYSTROBIN","BENZOVINDIFLUPYR","BOSCALID","CYPRODINIL",
                           "DIMETHOMORPH","FLUOPICOLIDE","METALAXYL","METCONAZOLE",
                           "MYCLOBUTANIL","PROPICONAZOLE","PYRACLOSTROBIN","PYRIMETHANIL",
                           "TEBUCONAZOLE","TETRACONAZOLE","TRIFLOXYSTROBIN"]]
adjuvant = pesticides_df["PIPERONYL BUTOXIDE"]
```

```python
#creating a new column into the pesticides_df referencing the created variables above
pesticides_df['Insecticide Sum'] = insecticides.sum(axis=1)
pesticides_df['Herbicide Sum'] = herbicides.sum(axis=1)
pesticides_df['Fungicide Sum'] = fungicides.sum(axis=1)
pesticides_df['Adjuvant Sum'] = adjuvant
pesticides_df.head()
```

```python
pesticides_df = pesticides_df.rename(columns = {'AREA_FIPS_CODE': 'FIPS'})
pesticides_df
```

```python
reduced_pesticides = pesticides_df[['FIPS', 'Insecticide Sum', 'Herbicide Sum', 'Fungicide Sum', 'Adjuvant Sum']]
reduced_pesticides
```

```python
reduced_pesticides.dtypes
```

```python
#calling in the cancer csv
cancer_df = pd.read_csv(Path("cancer_cleaned.csv"))
cancer_df.head()
```

```python
#looking at the columns of the cancer df
cancer_df.columns
```

```python
#reducing the columns to only the ones that we want to look at 
reduced_cancer = cancer_df[["County", " FIPS", "Age-Adjusted Incidence Rate([rate note]) - cases per 100,000"]]
reduced_cancer.head()
```

```python
#renaming the columns to be more legible/short
renamed_cancer = reduced_cancer.rename(columns = {' FIPS': 'FIPS', 'Age-Adjusted Incidence Rate([rate note]) - cases per 100,000':'Cancer Incidence Rate Per 100k'})
renamed_cancer.head(30)
```

```python
#looking at the tail of the dataframe since majority of the null values are at the end
renamed_cancer.tail(30)
```

```python
#dropping rows where 'data not available' , notice there was an extra space after the phrase ):
value_to_drop = 'data not available '
renamed_cancer = renamed_cancer.drop(renamed_cancer[renamed_cancer['Cancer Incidence Rate Per 100k'] == value_to_drop].index)
renamed_cancer.tail()
```

```python
#dropping the null values in the cancer df
renamed_cancer = renamed_cancer.dropna(how = 'any')
renamed_cancer.tail()
```

```python
#dropping the rows where theres a * (notice there was an extra space after the *)
value_to_drop = '* '
renamed_cancer = renamed_cancer.drop(renamed_cancer[renamed_cancer['Cancer Incidence Rate Per 100k'] == value_to_drop].index)
renamed_cancer.tail()
```

```python
renamed_cancer.head()
```

```python
#checking the dtypes so we can perform the table merge, the FIPS column was a float before but needed to be made
#into a string(object)
renamed_cancer.dtypes
```

```python
#converting the fips column from the cancer table into a a string(object)
renamed_cancer = renamed_cancer.astype({'FIPS': str}, errors = 'raise')
```

```python
#ensuring the data conversion was correct
renamed_cancer['FIPS'].dtype
```

```python
#changing the cancer incidence rate column into a float in case we need to do calculations with it
renamed_cancer = renamed_cancer.astype({'Cancer Incidence Rate Per 100k': float}, errors = 'raise')
```

```python
#checking the datatype of the incidence rate column to ensure it was correctly converted into a float
renamed_cancer['Cancer Incidence Rate Per 100k'].dtype
```

```python
#checking the cancer dataframe again
renamed_cancer.head()
```

```python
#checking the pesticides dataframe again
reduced_pesticides.head()
```

```python
#removing the trailing 0 thats in the FIPS code column in the cancer dataframe
renamed_cancer['FIPS'] = renamed_cancer['FIPS'].astype(str).str.rstrip('0').str.rstrip('.')
renamed_cancer.head()
```

```python
#merging the dataframes now that the datatypes and common column matches up
merged_df = pd.merge(renamed_cancer, reduced_pesticides, on = 'FIPS')
merged_df.head(100)
```

```python
#checking the dataframe info just to see how many rows were maintained after the merge
merged_df.info
```
