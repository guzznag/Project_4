```python
import pandas as pd
import sqlite3
from pathlib import Path
```

```python
# Connection to sqlite database
conn = sqlite3.connect("data/ncss_labdata.sqlite")
```

```python
# Explore database
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()

for table in tables:
    print(table[0])
```

```python
# Explore tables 4 - lab_site
# ** lat and long to find city, then map to county **
query4 = "SELECT * FROM lab_site"
df4 = pd.read_sql_query(query4, conn)
df4.head()
#df4.to_csv('long_lat.csv',index=False)
```

```python
# List columns in df4
df4.columns
```

```python
# Extract only long, lat, site id & rename
long_lat_df = df4[['user_site_id', 'site_key', 'latitude_std_decimal_degrees', 'longitude_std_decimal_degrees']]
long_lat_df = long_lat_df.rename(columns = {'user_site_id': 'site_id', 'site_key': 'site_key',
                                            'latitude_std_decimal_degrees': 'latitude', 
                                            'longitude_std_decimal_degrees': 'longitude'})
long_lat_df
```

```python
# Explore tables 8 - lab_mineralogy_glass_count
query8 = "SELECT * FROM lab_mineralogy_glass_count"
df8 = pd.read_sql_query(query8, conn)
df8.head()
```

```python
# List columns in df8
df8.columns
```

```python
# Explore tables 9 - lab_major_and_trace_elements_and_oxides
query9 = "SELECT * FROM lab_major_and_trace_elements_and_oxides"
df9 = pd.read_sql_query(query9, conn)
df9.head()
```

```python
# List columns in df9
df9.columns
```

```python
# Extract sample number, various elements results & rename
results_df = df9[['labsampnum', 'cadmium_trace_element', 'arsenic_trace_element','lead_trace_element', 
                  'barium_trace_element','chromium_trace_element', 'mercury_trace_element', 
                  'silver_trace_element', 'selenium_trace_element']]
results_df = results_df.rename(columns = {'labsampnum': 'sample_num', 'cadmium_trace_element': 'cadmium', 
                                          'arsenic_trace_element': 'arsenic','lead_trace_element': 'lead', 
                                          'barium_trace_element': 'barium', 
                                          'chromium_trace_element': 'chromium', 
                                          'mercury_trace_element': 'mercury', 
                                          'silver_trace_element': 'silver', 
                                          'selenium_trace_element': 'selenium'})
results_df
```

```python
# Explore tables 10 - lab_chemical_properties
query10 = "SELECT * FROM lab_chemical_properties"
df10 = pd.read_sql_query(query10, conn)
df10.head()
```

```python
# List columns in df10
df10.columns
```

```python
# Extract sample num and carbon/nitrogen ratio & rename
ratio_df = df10[['labsampnum', 'carbon_to_nitrogen_ratio']]
ratio_df = ratio_df.rename(columns = {'labsampnum': 'sample_num', 'carbon_to_nitrogen_ratio': 
                                      'carbon_to_nitrogen'})
ratio_df
```

```python
# Explore tables 11 - lab_calculations_including_estimates_and_default_values
query11 = "SELECT * FROM lab_calculations_including_estimates_and_default_values"
df11 = pd.read_sql_query(query11, conn)
df11.head()
```

```python
# List columns in df11
df11.columns
```

```python
# Explore tables 12 - lab_combine_nasis_ncss
# ** this table connects results tables to df4 which has lat and long **
query12 = "SELECT * FROM lab_combine_nasis_ncss"
df12 = pd.read_sql_query(query12, conn)
df12.head()
```

```python
# List columns in df12
df12.columns
```

```python
# Extract site id, site ket, sample num & rename
# Use as table to connect long/lat to results tables
table_link_df = df12[['site_key', 'pedlabsampnum', 'upedonid']]
table_link_df = table_link_df.rename(columns = {'site_key': 'site_key', 'pedlabsampnum': 'sample_num', 
                                                'upedonid': 'site_id'})
table_link_df
#table_link_df.to_csv('link table.csv')
```

```python
# Merge results tables
all_results_df = pd.merge(results_df, ratio_df, on='sample_num')
all_results_df
#all_results_df.to_csv('results.csv')
```

```python
trim_all_results_df = pd.read_csv(Path("results.csv"))
trim_all_results_df.head()
```

```python
# Add site id to results
results_with_site_df = pd.merge(table_link_df, trim_all_results_df, on='sample_num')
results_with_site_df
```

```python
# Combine results with long/lat
final_df = pd.merge(results_with_site_df, long_lat_df, on='site_id')
final_df = final_df.fillna(0)
final_df = final_df[final_df['latitude'] != 0]
final_df
#final_df.to_csv('final test results.csv')
```
