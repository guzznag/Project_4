```python
#Import Dependencies
import pandas as pd 
from pathlib import Path
from sklearn.metrics import confusion_matrix, classification_report 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import hvplot.pandas
```

```python
#Read in CSV
data = pd.read_csv('all_cancer_and_pesticides.csv')

#Display Data
data 
```

```python
value_to_drop = 'Union County, Florida(6)'
data = data.drop(data[data['County'] == value_to_drop].index)
data
```

```python
#Drop all columns excepts, FIPS, Incident Rate Per 100K, and pesticide columns
updated_data = data[["FIPS", "Cancer Incidence Rate Per 100k", "Insecticide Sum", "Herbicide Sum", "Fungicide Sum"]]
updated_data
```

```python
# Separate the y variable, the labels
y = updated_data['Cancer Incidence Rate Per 100k']

# Separate the X variable, the features
X = updated_data.drop(columns='Cancer Incidence Rate Per 100k')
```

```python
#Define the model
model = LinearRegression()
model.fit(X,y)
```

```python
#Find the best fit line
print(f"Model's formula: y = {model.intercept_} + {model.coef_[0]}X")
```

```python
#Calculate predicted y
predicted_y_values = model.predict(X)
```

```python
#Evaluate the Model
# Compute metrics for the linear regression model: score, r2, mse, rmse, std
score = model.score(X, y, sample_weight=None)
r2 = r2_score(y, predicted_y_values)
mse = mean_squared_error(y, predicted_y_values)
rmse = np.sqrt(mse)
std = np.std(y)
# Print relevant metrics.
print(f"The score is {score}.")
print(f"The r2 is {r2}.")
print(f"The mean squared error is {mse}.")
print(f"The root mean squared error is {rmse}.")
print(f"The standard deviation is {std}.")
```
