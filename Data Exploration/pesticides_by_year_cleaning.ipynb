```python
import pandas as pd
from pathlib import Path
```

```python
pesticides_2019 = pd.read_csv(Path("2019_pesticide_by_county.csv"))
pesticides_2019.head()
```

```python
insecticides = pesticides_2019[["ACEPHATE", "ACETAMIPRID", "CARBARYL","CHLORANTRANILIPROLE",
                              "CHLORPYRIFOS","CLOTHIANIDIN","CYANTRANILIPROLE",
                              "DIAZINON","DICROTOPHOS","DIFLUBENZURON","DIMETHOATE",
                              "DINOTEFURAN","ETHOPROPHOS","ETOXAZOLE","FIPRONIL",
                              "IMIDACLOPRID","MALATHION","METHOMYL","METHOXYFENOZIDE",
                              "PERMETHRIN","SULFOXAFLOR","TEBUPIRIMPHOS","TERBUFOS",
                              "THIAMETHOXAM"]]
herbicides = pesticides_2019[["2,4-D","ACETOCHLOR","ATRAZINE","BENTAZONE",
                            "BROMACIL","BROMOXYNIL","CHLORIMURON","DICAMBA",
                            "DIMETHENAMID","DIMETHENAMID & DIMETHENAMID-P",
                            "DIMETHENAMID-P","DIURON","FLUOMETURON","GLYPHOSATE",
                           "HALOSULFURON","HEXAZINONE","IMAZETHAPYR","LINURON","METOLACHLOR",
                           "METOLACHLOR & METOLACHLOR-S","METOLACHLOR-S","METRIBUZIN",
                           "ORYZALIN","PROPAZINE","SIMAZINE","SULFENTRAZONE","TEBUTHIURON",
                           "THIOBENCARB","TRICLOPYR"]]
fungicides = pesticides_2019[["AZOXYSTROBIN","BENZOVINDIFLUPYR","BOSCALID","CYPRODINIL",
                           "DIMETHOMORPH","FLUOPICOLIDE","METALAXYL","METCONAZOLE",
                           "MYCLOBUTANIL","PROPICONAZOLE","PYRACLOSTROBIN","PYRIMETHANIL",
                           "TEBUCONAZOLE","TETRACONAZOLE","TRIFLOXYSTROBIN"]]
```

```python
pesticides_2019['Insecticide Sum'] = insecticides.sum(axis=1)
pesticides_2019['Herbicide Sum'] = herbicides.sum(axis=1)
pesticides_2019['Fungicide Sum'] = fungicides.sum(axis=1)
pesticides_2019.head()
```

```python
pesticides_2019.tail()
```

```python
insecticide_total = pesticides_2019["Insecticide Sum"].sum
herbicide_total = pesticides_2019["Herbicide Sum"].sum
fungicide_total = pesticides_2019["Fungicide Sum"].sum
```
