```python
import pandas as pd
from pathlib import Path
```

```python
#calling in the 2019 pesticides table
pesticides_df = pd.read_csv(Path("../Data Sets/Raw Sets/2019_pesticide_by_country.csv"))
pesticides_df.head()
```

```python
#filling in the null values with a 0
pesticides_df = pesticides_df.fillna(0)
pesticides_df.head()
```

```python
#looking at each of the different column names/dataframe info
pesticides_df.info
```

```python
#looking at each column name
pesticides_df.columns
```

```python
#dropping the grand total column since we don't really need it
pesticides_df = pesticides_df.drop(columns = "Grand Total")
pesticides_df
```

```python
#creatng variables to hold specific columns that fall within that category 
insecticides = pesticides_df[["ACEPHATE", "ACETAMIPRID", "CARBARYL","CHLORANTRANILIPROLE",
                              "CHLORPYRIFOS","CLOTHIANIDIN","CYANTRANILIPROLE",
                              "DIAZINON","DICROTOPHOS","DIFLUBENZURON","DIMETHOATE",
                              "DINOTEFURAN","ETHOPROPHOS","ETOXAZOLE","FIPRONIL",
                              "IMIDACLOPRID","MALATHION","METHOMYL","METHOXYFENOZIDE",
                              "PERMETHRIN","SULFOXAFLOR","TEBUPIRIMPHOS","TERBUFOS",
                              "THIAMETHOXAM"]]
herbicides = pesticides_df[["2,4-D","ACETOCHLOR","ATRAZINE","BENTAZONE",
                            "BROMACIL","BROMOXYNIL","CHLORIMURON","DICAMBA",
                            "DIMETHENAMID","DIMETHENAMID & DIMETHENAMID-P",
                            "DIMETHENAMID-P","DIURON","FLUOMETURON","GLYPHOSATE",
                           "HALOSULFURON","HEXAZINONE","IMAZETHAPYR","LINURON","METOLACHLOR",
                           "METOLACHLOR & METOLACHLOR-S","METOLACHLOR-S","METRIBUZIN",
                           "ORYZALIN","PROPAZINE","SIMAZINE","SULFENTRAZONE","TEBUTHIURON",
                           "THIOBENCARB","TRICLOPYR"]]
fungicides = pesticides_df[["AZOXYSTROBIN","BENZOVINDIFLUPYR","BOSCALID","CYPRODINIL",
                           "DIMETHOMORPH","FLUOPICOLIDE","METALAXYL","METCONAZOLE",
                           "MYCLOBUTANIL","PROPICONAZOLE","PYRACLOSTROBIN","PYRIMETHANIL",
                           "TEBUCONAZOLE","TETRACONAZOLE","TRIFLOXYSTROBIN"]]
adjuvant = pesticides_df["PIPERONYL BUTOXIDE"]
```

```python
#creating a new column into the pesticides_df referencing the created variables above
pesticides_df['Insecticide Sum'] = insecticides.sum(axis=1)
pesticides_df['Herbicide Sum'] = herbicides.sum(axis=1)
pesticides_df['Fungicide Sum'] = fungicides.sum(axis=1)
pesticides_df['Adjuvant Sum'] = adjuvant
pesticides_df.head()
```

```python
pesticides_df = pesticides_df.rename(columns = {'AREA_FIPS_CODE': 'FIPS'})
pesticides_df
```

```python
reduced_pesticides = pesticides_df[['FIPS', 'Insecticide Sum', 'Herbicide Sum', 'Fungicide Sum', 'Adjuvant Sum']]
reduced_pesticides
```
