```python
def find_region_with_highest_isolates(data):
    region_pathogen_counts = {}  # Dictionary to store counts of isolates for each region and pathogen

    # Iterate through the dataset to count isolates for each region and pathogen
    for entry in data:
        region = entry['HHS_Region']
        pathogen = entry['Pathogen']
        isolates = entry['Number_of_isolates']
        
        # Initialize count for the region if it doesn't exist
        if region not in region_pathogen_counts:
            region_pathogen_counts[region] = {}
        
        # Initialize count for the pathogen in the region if it doesn't exist
        if pathogen not in region_pathogen_counts[region]:
            region_pathogen_counts[region][pathogen] = 0
        
        # Increment count of isolates for the region and pathogen
        region_pathogen_counts[region][pathogen] += isolates

    # Find the region with the highest number of isolates
    max_region = max(region_pathogen_counts, key=lambda x: sum(region_pathogen_counts[x].values()))
    max_isolates = sum(region_pathogen_counts[max_region].values())

    # Find the pathogen for the region with the highest number of isolates
    max_pathogen = max(region_pathogen_counts[max_region], key=region_pathogen_counts[max_region].get)

    return max_region, max_pathogen, max_isolates

# Example usage:
data = [
    {'HHS_Region': 'Region 1 - Boston', 'Pathogen': 'Campylobacter', 'Number_of_isolates': 26},
    {'HHS_Region': 'Region 1 - Boston', 'Pathogen': 'Campylobacter', 'Number_of_isolates': 12},
    {'HHS_Region': 'Region 9 - San Francisco', 'Pathogen': 'Vibrio', 'Number_of_isolates': 3},
    # Add more data entries as needed...
]

region, pathogen, isolates = find_region_with_highest_isolates(data)
print("Region with the highest number of isolates:", region)
print("Pathogen:", pathogen)
print("Number of isolates:", isolates)
```
